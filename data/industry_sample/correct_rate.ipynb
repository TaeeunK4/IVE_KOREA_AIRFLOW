```python
import pandas as pd

df_true = pd.read_csv('ive_industry_sample_true.csv', encoding='euc-kr')
df_result_1 = pd.read_csv('ive_industry_sample_result_1.csv', encoding='euc-kr')
df_result_2 = pd.read_csv('ive_industry_sample_result_2.csv', encoding='euc-kr')
df_result_3 = pd.read_csv('ive_industry_sample_result_3.csv', encoding='euc-kr')
```

```python
df_1 = df_true.merge(df_result_1, how='inner', on='ADS_IDX')

# 3. 정답 여부 체크
df_1['IS_CORRECT'] = df_1['INDUSTRY'] == df_1['TRUE_INDUSTRY']

# 4. 통계 계산
total_count = len(df_1)
correct_count = df_1['IS_CORRECT'].sum()
accuracy = (correct_count / total_count) * 100 if total_count > 0 else 0

print("-" * 30)
print(f"전체 평가 데이터 수: {total_count}")
print(f"정답 수: {correct_count}")
print(f"정답률(Accuracy): {accuracy:.2f}%")
print("-" * 30)

# 5. 오답 리스트 추출 (분석용)
incorrect_df = df_1[df_1['IS_CORRECT'] == False]

if not incorrect_df.empty:
    print("\n[오답 리스트]")
    # NAME, 예측값, 실제값 위주로 출력
    print(incorrect_df[['ADS_IDX', 'NAME_x', 'INDUSTRY', 'TRUE_INDUSTRY']])
```
